# Histograms

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sns.set(color_codes=True)
```

```python
x = np.random.normal(size=100)
sns.distplot(x);
```

```python
the line is the kernel density estimate (KDE)
```

Now, remove the kde and add a rugplot (where each tick is an observation)

```python
sns.distplot(x, kde=False, rug=True);
```

# Modifying the number of bins

```python
sns.distplot(x, bins=20, kde=False, rug=True);
```

# Histogram with 2 variables

```python
import seaborn as sns
df = sns.load_dataset('iris')
 
# Method 1: on the same Axis
sns.distplot( df["sepal_length"] , color="skyblue", label="Sepal Length")
sns.distplot( df["sepal_width"] , color="red", label="Sepal Width")
```

# Grouped boxplots and stripplot

This example is from:

https://cmdlinetips.com/2019/03/how-to-make-grouped-boxplots-in-python-with-seaborn/

Read-in a dataset example:

```python
data_url = 'http://bit.ly/2cLzoxH'
# read data from url as pandas dataframe
gapminder = pd.read_csv(data_url)
print(gapminder.head(3))
```

```python
# subset the gapminder data frame for rows with year values 1952 and 2007
df1 = gapminder[gapminder['year'].isin([1952,2007])]
# get a look at the data with head function
df1.head(n=3)
```

## Grouped boxplot

Let us make a grouped boxplot with continent on x-axis and lifeExp on the y-axis such that we see distributions of lifeExp for two years separately for each continent.

```python
sns.boxplot(y='lifeExp', x='continent', 
                 data=df1, 
                 palette="colorblind",
                 hue='year')
```

## Grouped stripplot

```python
bplot=sns.stripplot(y='lifeExp', x='continent', 
                   data=df1, 
                   jitter=True, 
                   marker='o', 
                   alpha=0.5,
                   hue='year')
```

The dodge=True argument splits the data nicely like you see in grouped boxplot and each year in different color.

```python
sns.stripplot(y='lifeExp', x='continent', 
                   data=df1, 
                   jitter=True,
                   dodge=True,
                   marker='o', 
                   alpha=0.5,
                   hue='year')
```

## Plotting jitter on a Boxplot (i.e. observations on the same boxplot) 

```python
# library & dataset
import seaborn as sns

%matplotlib inline

df = sns.load_dataset('iris')
 
# Usual boxplot
ax = sns.boxplot(x='species', y='sepal_length', data=df)
 
# Add jitter with the swarmplot function.
ax = sns.swarmplot(x='species', y='sepal_length', data=df, color="grey")
```
