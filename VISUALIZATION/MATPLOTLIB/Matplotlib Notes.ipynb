# Scatter plot of one numerical variable vs another one coloured by a categorical var

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

# Plotting by the categories of a variable using seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd

carat = [5, 10, 20, 30, 5, 10, 20, 30, 5, 10, 20, 30]
price = [100, 100, 200, 200, 300, 300, 400, 400, 500, 500, 600, 600]
color =['D', 'D', 'D', 'E', 'E', 'E', 'F', 'F', 'F', 'G', 'G', 'G',]

df = pd.DataFrame(dict(carat=carat, price=price, color=color))

sns.lmplot('carat', 'price', data=df, hue='color', fit_reg=False)

plt.show()
```

## Plotting using the index as the X plot

```python
df = pd.DataFrame(dict(price=price, color=color))
df=df.reset_index()
print(df)
```

```python
sns.lmplot('index','price', data=df, hue='color', fit_reg=False)
```

## Fontsize

```python
import matplotlib.pyplot as plt

plt.figure()
x = [1,2]
y = [1,2]
plt.plot(x,y)
plt.xlabel('x-axis', fontsize=20)
# specify font sizes
plt.ylabel('y-axis', fontsize=20)
plt.suptitle('Graph Title', fontsize=30)
```

## Venn diagrams

* 2-circles

```python
from matplotlib_venn import venn2

venn2([set(['A', 'B', 'C', 'D']), set(['D', 'E', 'F'])])
```

```python
venn2(subsets = (3, 2, 1))
```

Let's add a title

```python
from matplotlib_venn import venn2

import matplotlib.pyplot as plt

venn2(subsets = (3, 2, 1))

plt.title("I made a Venn Diagram!");
```

* Unweighted

```python
from matplotlib_venn import venn2_unweighted

venn2_unweighted(subsets = (3, 2, 1))
```

Adding the comma-separation for thousand

```python
venn2_unweighted(subsets = (format(3000,','), format(2000,','), format(1000,',')))
```

* 3-circles

## Horizontal lines

The list : [1, 3, 5] prints the horizontal lines in these Y-axis coords. -3, 5 means from where to where the line will be plotted

```python
import numpy as np 
from matplotlib import patches 
import matplotlib.pyplot as plt 
   
fig, ax = plt.subplots() 
ax.hlines([1, 3, 5], -3, 5, color ="green") 
ax.set_title('matplotlib.axes.Axes.hlines Example') 
  
plt.show() 
```

```python
fig, ax = plt.subplots() 
ax.hlines([1, 3, 5], color ="green") 
ax.set_title('matplotlib.axes.Axes.hlines Example') 
  
plt.show() 
```
