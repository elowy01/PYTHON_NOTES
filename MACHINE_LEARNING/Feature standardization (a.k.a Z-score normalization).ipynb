# Background
Feature standardization is the process by which the different predictors or features are transformed so features that may have different units are comparable. This normalization will transform the different variables so all them will have an average=0 and a standard deviation=1, which are the properties of a standard normal distribution. This properties are important because there are a lot of machine learning algorithms that assume that the different features are normally distributed and that have the same units.  
Also, standardization is useful if you have features that have different ranges (i.e. one feature is in the ranges of 1000s and other in 10s)

# Standardization
`sklearn` as a useful library named `preprocessing` that can be used to perform the standardization. Before using this function, let's convert the `iris_data` into a Pandas dataframe

```python
import numpy as np

DF = pd.DataFrame(data= np.c_[iris_data['data'], iris_data['target']],
                  columns= iris_data['feature_names'] + ['target'])
```

Now, we use the preprocessing library:

```python
from sklearn import preprocessing

std_scale = preprocessing.StandardScaler().fit(DF[iris_data['feature_names']])
DF_std = std_scale.transform(DF[iris_data['feature_names']])
```

Now, we can check the mean and std for the transformed variables

```python
print("sepal length (cm)-Mean: {0}".format(DF_std[:,0].mean()))
print("sepal width (cm)-Mean: {0}".format(DF_std[:,1].mean()))
print("petal length (cm)-Mean: {0}".format(DF_std[:,2].mean()))
print("petal width (cm)-Mean: {0}".format(DF_std[:,2].mean()))
print("sepal length (cm)-Std: {0}".format(DF_std[:,0].std()))
print("sepal width (cm)-Std: {0}".format(DF_std[:,1].std()))
print("petal length (cm)-Std: {0}".format(DF_std[:,2].std()))
print("petal width (cm)-Std: {0}".format(DF_std[:,2].std()))
```

```python
from pandas.plotting import scatter_matrix

%matplotlib inline

scatter_matrix(DF[iris_data['feature_names']], figsize=(12, 8), color="red")
```

Now, let's create a scatter plot matrix with the normalized features.

```python
aDF_std=pd.DataFrame(data=DF_std,columns=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'])
scatter_matrix(aDF_std[iris_data['feature_names']], figsize=(12, 8),color="green")
```
